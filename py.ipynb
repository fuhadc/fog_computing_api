```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_dynamic_graph(csv_path):
    # Read data from CSV into a pandas DataFrame
    df = pd.read_csv(csv_path)
    
    # Loop through columns to determine data types and characteristics
    sns.set(style="ticks")
    sns.pairplot(df, diag_kind="kde")
    plt.show()
    
            

csv_file_path = "/Users/muhammedfuhadc/Desktop/cloud_api/uploads/data/DSZdBdP1wP1stFE39lGW0jUQ5uGyzN4A/webrtc.csv"
plot_dynamic_graph(csv_file_path)
```
